```python
# === YOUR DATA ===
# A housing society collected these prices (in lakhs)
# Some data has errors — negative prices, None values, zeros
import math

house_prices = [45, 62, None, 38, -5, 71, 0, 55, None, 89, 42, 0, 63, 77, -12, 58]
house_sizes  = [900, 1200, 1100, 800, 1500, 1400, None, 1050, 950, 1800, 850, 700, 1250, 1600, 1300, 1100]
locations    = ["Delhi", "Mumbai", "Delhi", "Pune", "Mumbai", "Delhi",
                "Pune", "Mumbai", "Delhi", "Mumbai", "Pune", "Delhi",
                "Mumbai", "Delhi", "Pune", "Mumbai"]

# =========================================================
# TASK 1 — Clean the data
# Remove entries where price is None, 0, or negative
# (remove the matching size and location too)
# =========================================================

clean_prices    = []
clean_sizes     = []
clean_locations = []

clean_prices = list(filter(lambda x: x is not None and x!=0 and x>0,house_prices))


print("Cleaned entries:", len(clean_prices))
# Expected: 11 entries

clean_sizes = list(filter(lambda x: x is not None, house_sizes))

print("Cleaned sizes:", len(clean_sizes))

clean_locations = list(filter(lambda x: x is not None, locations))

print("Clean locations:", len(clean_locations))

# =========================================================
# TASK 2 — Basic statistics (without using any library)
# Calculate mean, max, min, and range of clean prices
# =========================================================

mean_price = None  
mean_price = sum(clean_prices)/len(clean_prices)
max_price  = None 
max_price = max(clean_prices)
min_price  = None 
min_price = min(clean_prices)
price_range = None
price_range = max_price - min_price

print(f"Mean:  {mean_price:.1f} lakhs")
print(f"Max:   {max_price} lakhs")
print(f"Min:   {min_price} lakhs")
print(f"Range: {price_range} lakhs")


# =========================================================
# TASK 3 — Group by location
# Find the average price for each city
# =========================================================

city_prices = {}   # { "Delhi": [list of prices], ... }

for i in range(len(clean_prices)):
    city = clean_locations[i]
    prices = clean_prices[i]
    pass

# Now compute average per city
for city, prices in city_prices.items():
    avg = sum(prices) / len(prices)
    print(f"{city}: ₹{avg:.1f} lakhs avg")


# =========================================================
# TASK 4 — Price per square foot
# Create a new list: price_per_sqft for each clean entry
# Formula: (price in lakhs × 100000) / size in sqft
# =========================================================

price_per_sqft = {}

for i in range(len(clean_prices)):
    city = clean_locations[i]
    price = clean_prices[i]
    size = clean_sizes[i]
    
    price_per_sqft[city] = (price*100000) / size

top_3  = sorted(price_per_sqft.items(),key = lambda x:x[1], reverse = True) [:3]
print("3 Most expensive per sqft are:")
for city,value in top_3:
    print(city, ":", value)

# Print the top 3 most expensive per sqft
# Hint: use sorted() with a key


# =========================================================
# TASK 5 — Find outliers (simple method)
# Any price more than 1.5× the mean is an outlier
# Print those prices and their locations
# =========================================================

print("\nOutliers:")
for i in range(len(clean_prices)):
    price = clean_prices[i]
    city = clean_locations[i]
    
    if price > 1.5*mean_price:
        print(city, ":", price, "lakhs")
        
        
```
