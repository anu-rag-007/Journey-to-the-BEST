```python
# 1. What % of passengers survived overall?
# 2. What was the average fare per class?
# 3. How many passengers embarked from each port?
# 4. What was the youngest and oldest passenger per class?
# 5. Create a new column: AgeGroup (Child/Teen/Adult/Senior)
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv(r'C:\Users\Hp\.vscode\train.csv')
total = df.PassengerId.count()
survived = df[df['Survived']==1]['Survived'].count()
print(f"{((survived/total)*100).round(2)} % of Passengers Survived Overall")
```

```python
avg_fare = df.groupby('Pclass')['Fare'].mean()
print(avg_fare)
```

```python
embarked = df.groupby('Embarked')['Embarked'].count()
print(embarked)
```

```python
youngest = df.groupby('Pclass')['Age'].min()
oldest = df.groupby('Pclass')['Age'].max()

print("Youngest:\n", youngest*12, "\nOldest:\n", oldest.astype(int))
```

```python
conditions = [
    (df['Age'] < 13),
    (df['Age'] >= 13) & (df['Age'] < 18),
    (df['Age'] >= 18) & (df['Age'] < 60),
    (df['Age'] >= 60)
]

values = ['Child', 'Teen', 'Adult', 'Senior']

df['Agegroup'] = np.select(conditions, values, default = 'unknown')
df.to_csv(r'C:\Users\Hp\.vscode\train.csv')
```
