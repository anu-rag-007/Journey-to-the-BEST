```python
import numpy as np

prices = np.array([45,62,38,71,55,89,42,63,77,58], dtype=float)
sizes  = np.array([900,1200,800,1400,1050,1800,850,1250,1600,1100], dtype=float)

#  We have to find the mean, variance and standard variance without using np.mean, np.std
#  We can use only np.sum, np.sqrt & len()
```

```python
#  MEAN OF PRICES & SIZES

n = len(prices)  # And of sizes
mean_p = np.sum(prices)/n
mean_s = np.sum(sizes)/n
print("Mean: ", mean_p)
```

```python
#  VARIANCE OF PRICES & SIZES

for i in range(n):
    var_p = np.sum((prices[i] - mean_p)**2)/n
    var_s = np.sum((prices[i] - mean_s)**2)/n
    
print(var_p)
```

```python
#  STANDARD DEVIATION OF PRICES & SIZES

std_p = np.sqrt(var_p)
std_s = np.sqrt(var_s)
print(std_p)
```

```python
#   CORRELATION B/W PRICES AND SIZES

for i in range(n):
   diff_p = prices[i] - mean_p
   diff_s = prices[i] - mean_s
cov = np.sum(diff_p*diff_s)/n
corr = cov / (std_p * std_s)
print(corr)
```

```python
# Verify with numpy built-ins

print("Your mean:", mean_p, "| np.mean:", np.mean(prices))
print("Correlation:", round(corr, 3))
```
