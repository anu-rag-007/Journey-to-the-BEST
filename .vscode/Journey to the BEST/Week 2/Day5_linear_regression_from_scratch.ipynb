```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset: house size → price
X = np.array([800,1000,1200,1500,1800,2000,2200,2500], dtype=float)
y = np.array([40, 50,  58,  72,  85,  95, 102, 118], dtype=float)
```

```python
X_Norm = np.mean(X - X.mean())/X.std()

np.random.seed(42)
w = np.random.randn()   #intercept
b = np.random.randn()   #slope

lr  = 0.01
epochs = 1000
losses = []

for epoch in range(epochs):
    y_hat = (w*X_Norm) + b
    loss = np.mean((y_hat - y)**2)
    
    dw = np.mean(2*(y_hat - y)*(X_Norm))
    db = np.mean(2*(y_hat - y))
    
    #  Updated weights
    w = w - lr*dw
    b = b - lr*db
    
    losses.append(loss)
    
    
print(f"Final loss: {losses[-1]:.2f}")
print(f"w={w:.3f}, b={b:.3f}")

# Plot loss curve
plt.plot(losses)
plt.title("Loss over epochs")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.show()
    
    
```
