```python
import numpy as np
import math as m

def sigmoid(x):
    return 1/(1+m.e**(-x))

def softmax(x):
    return max(x) - m.e**x/sum(m.e**x)
```

```python
x1 = np.array([-2, -1, 0, 1, 2])
x2 = np.array([2.0, 1.0, 0.1])

print("Sigmoid:", sigmoid(x1).round(3))

print("Softmax:", softmax(x2).round(3))
print("Sum:", softmax(x2).sum())
```
